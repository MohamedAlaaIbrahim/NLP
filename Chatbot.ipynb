```python
# -*- coding: utf-8 -*-
"""
Created on Thu Jun  3 16:05:35 2021

@author: Omar
"""

import nltk
from nltk import pos_tag,TreebankWordTokenizer
import numpy as np
import os
import csv
import pandas as pd
```

```python
headers = []
body = []
with open('train_arabic.csv', 'r',encoding='utf-16-le') as reader:
    lines = reader.readlines()
    count = 0
    for line in lines:
        if(count%10000 ==0):
            print("Finished "+str(count)+ "lines")
        if(count == 0):
            headers.append(line.split(','))
        else:
            body.append(line.split(','))

        count += 1

df_with_empty_col = pd.DataFrame(data = body,columns = headers)
```

```python
df = df_with_empty_col.iloc[:,:-3]
df
```
